```python
#奇摩電影搜尋
import requests
from bs4 import BeautifulSoup
url = 'https://movies.yahoo.com.tw/movie_thisweek.html'
r = requests.get(url)
soup = BeautifulSoup(r.text)
infos = []
res = []
for page in range(1, 5):
    url = 'https://movies.yahoo.com.tw/movie_intheaters.html?page=%s' % page
    r = requests.get(url)
    soup = BeautifulSoup(r.text)
    for d in soup.find_all('div', class_="release_info"):
        url2 = d.find('div', class_="release_movie_name").find('a', class_='gabtn')['href']
        r2 = requests.get(url2)
        soup2 = BeautifulSoup(r2.text)
        info = {}
        info['名稱'] = soup2.find('div', class_="movie_intro_info_r").find('h1', ).text
        info['分類'] = [d.find('a', class_='gabtn').text.replace(' ', '').replace('\n', '') for d in soup2.find('div', class_="level_name_box").find_all('div', class_="level_name")]
        info['劇情介紹'] = soup2.find('div', class_="gray_infobox_inner").find('span', ).text.replace('\n','').replace('\r','').replace('\xa0','').replace(' ','')
        info['上映日期'] = soup2.find('div', class_="movie_intro_info_r").find('span', ).text.replace('上映日期：', '')
        infos.append(info)
infos
```
