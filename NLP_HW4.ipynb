```python
import json
import os

def readdata():
  with open('wiki_text.txt', 'w', encoding='utf-8') as f:
    n = ['A','B','C','D','E','F','G','H','I','J','K','L','M']
    for i in n:
      dpath = '/content/drive/MyDrive/wiki_zh/A%s/' % i
      FileList = os.listdir(dpath)
      for nfile in FileList:
        Newfile = open(dpath + nfile, newline='')
        words = Newfile.readlines()
        for text in words:
            js = json.loads(text)
            f.write(' '+js['text'])

readdata()
```

```python
import jieba
from opencc import OpenCC


# Initial
cc = OpenCC('s2t')


def make_seg():
  with open('wiki_text_seg.txt', 'w', encoding='utf-8') as new_f:
      with open('wiki_text.txt', 'r', encoding='utf-8') as f:
          for data in f:
              data = cc.convert(data)
              data = jieba.cut(data)
              data = [word for word in data if word != ' ']
              data = ' '.join(data)
              new_f.write(data)

make_seg()
```

```python
from gensim.models import word2vec, fasttext


# Settings
seed = 666
sg = 0
window_size = 10
vector_size = 100
min_count = 1
workers = 8
epochs = 5
batch_words = 10000


# Train
train_data = word2vec.LineSentence('wiki_text_seg.txt')
model = fasttext.FastText(
    train_data,
    min_count=min_count,
    size=vector_size,
    workers=workers,
    iter=epochs,
    window=window_size,
    sg=sg,
    seed=seed,
    batch_words=batch_words,
)


model.save('fasttext.model')
```
